```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('./DogMoveData.csv')
data2 = pd.read_csv('./HPI_Datatable_FINAL.csv')
data3 = pd.read_csv('./Output_labeling_data_221129_FINAL.csv')
data4 = pd.read_csv('./dog_behavior_data_ver3.csv')
```

```python
data1['t_sec']=(data1['t_sec']*1000).tolist()
data1['t_sec']=pd.to_datetime(data1['t_sec'], unit='ms', origin='2022-10-01 16:00:00.000')
```

```python
data1=data1.iloc[::2, :] #짝수 인덱스 행만 추출
idx=[]

task_undefined=data1[(data1['Task']=='<undefined>')].index # Task가 <undefined> 인 인덱스 찾기

data1=data1.drop(task_undefined) #행 제거
data1=data1.reset_index(drop=True) #인덱스 초기화

data1.Task.value_counts()
```

```python
task_play=data1[(data1['Task']=='Task play')] # Task가 Task play 인 것만 추출

task_play_lying=(task_play[((task_play['Behavior_1']=='Lying chest') | (task_play['Behavior_2']=='Lying chest') | (task_play['Behavior_3']=='Lying chest'))].index).tolist() # lying chest 하나 이상 포함된 인덱스 찾기
task_play_eating=(task_play[((task_play['Behavior_1']=='Eating') | (task_play['Behavior_2']=='Eating') | (task_play['Behavior_3']=='Eating'))].index).tolist() # Eating 하나 이상 포함된 인덱스 찾기
task_play_panting=(task_play[((task_play['Behavior_1']=='Panting') | (task_play['Behavior_2']=='Panting') | (task_play['Behavior_3']=='Panting'))].index).tolist() # Panting 하나 이상 포함된 인덱스 찾기
task_play_pacing=(task_play[((task_play['Behavior_1']=='Pacing') | (task_play['Behavior_2']=='Pacing') | (task_play['Behavior_3']=='Pacing'))].index).tolist() # Pacing 하나 이상 포함된 인덱스 찾기
task_play_standing=(task_play[((task_play['Behavior_1']=='Standing') | (task_play['Behavior_2']=='Standing') | (task_play['Behavior_3']=='Standing'))].index).tolist() # Standing 하나 이상 포함된 인덱스 찾기
task_play_sitting=(task_play[((task_play['Behavior_1']=='Sitting') | (task_play['Behavior_2']=='Sitting') | (task_play['Behavior_3']=='Sitting'))].index).tolist() # Sitting 하나 이상 포함된 인덱스 찾기
task_play_sniffing=(task_play[((task_play['Behavior_1']=='Sniffing') | (task_play['Behavior_2']=='Sniffing') | (task_play['Behavior_3']=='Sniffing'))].index).tolist() # Sniffing 하나 이상 포함된 인덱스 찾기
task_play_bowing=(task_play[((task_play['Behavior_1']=='Bowing') | (task_play['Behavior_2']=='Bowing') | (task_play['Behavior_3']=='Bowing'))].index).tolist() # Bowing 하나 이상 포함된 인덱스 찾기

idx=task_play_lying+task_play_eating+task_play_panting+task_play_pacing+task_play_standing+task_play_sitting+task_play_sniffing+task_play_bowing # 찾은 인덱스를 하나의 리스트에 저장

data1=data1.drop(idx) #행 제거

data1=data1.reset_index(drop=True) #인덱스 초기화
```

```python
data1=data1.drop(labels=['TestNum', 'ABack_x', 'ABack_y', 'ABack_z', 'GBack_x', 'GBack_y', 'GBack_z', 'Task', 'Behavior_1', 'Behavior_2', 'Behavior_3', 'PointEvent'], axis=1) #열 제거

data1.columns=['Pname', 'Htime', 'Gyro_x', 'Gyro_y', 'Gyro_z', 'Acc_x', 'Acc_y', 'Acc_z'] #컬럼명 재정의

data1=data1[['Htime', 'Gyro_x', 'Gyro_y', 'Gyro_z', 'Acc_x', 'Acc_y', 'Acc_z', 'Pname']] #컬럼 순서 변경
```

```python
data3.drop('Index', axis=1, inplace=True)
data3.drop('filename', axis=1, inplace=True)
data3.drop('Class', axis=1, inplace=True)

data3 = data3.rename(columns={'Pname': '이름'})
data3 = data3.rename(columns={'starttime_ms': '시작시간(ms)'})
data3 = data3.rename(columns={'starttime': '시작시간'})
data3 = data3.rename(columns={'endtime_ms': '종료시간(ms)'}) 
data3 = data3.rename(columns={'endtime': '종료시간'})
data3 = data3.rename(columns={'pose': '자세'})
data3 = data3.rename(columns={'behavior': '행동'})
data3 = data3.rename(columns={'sound': '소리'})

data3['기타'] = np.nan

data3['자세'] = data3['자세'].replace({
    1:'서다',
    2:'뒷다리만 앉다',
    3:'모든 발이 앉다 (머리는 들고있음)', 
    4:'두 앞발을 들어올리다',
    5:'모든 발과 머리가 바닥과 닿은 채로 눕다',
    6:'옆으로 등을 대고 눕다',
    0:'해당없음'
})

data3['행동'] = data3['행동'].replace({
    8:'해당없음', 
    9:'걷다', 
    17:'먹다',                                                   
    13:'킁킁대다'
})

data3['소리'] = data3['소리'].replace({
    20:'해당없음', 
    21:'짖다', 
    23:'으르렁거리다'
})
```
